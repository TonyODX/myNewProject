```
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns;
```

```
df = pd.read_csv("/content/drive/MyDrive/LaLiga_24-25.csv");
print(df);
```

```
dfHT = pd.DataFrame(df, columns=["HomeTeam"]);
dfAT = pd.DataFrame(df, columns=["AwayTeam"]);
dfFTHG = pd.DataFrame(df, columns=["FTHG"]);
dfFTAG = pd.DataFrame(df, columns=["FTAG"]);
dfTYCHT = pd.DataFrame(df, columns=["HY"]);
dfTRCHT = pd.DataFrame(df, columns=["HR"]);
dfTYCAT = pd.DataFrame(df, columns=["AY"]);
dfTRCAT = pd.DataFrame(df, columns=["AR"]);
display(dfHT);
```

```
ind = list((dfHT[dfHT["HomeTeam"] == "Barcelona"].index));
ind2 = list((dfAT[dfAT["AwayTeam"] == "Barcelona"].index));
lstFTHG = df['FTHG'].values.tolist();
lstFTAG = df["FTAG"].values.tolist();
lstTYCHT = df["HY"].values.tolist();
lstTRCHT = df["HR"].values.tolist();
lstTYCAT = df["AY"].values.tolist();
lstTRCAT = df["AR"].values.tolist();
print(ind);
print(lstFTHG);
```

```
totGoalsHT = 0;
totGoalsAT = 0;
totalYCHT = 0;
totalYCAT = 0;
totalRCHT = 0;
totalRCAT = 0;
totMatches = 0;
for a in ind:
  totGoalsHT += lstFTHG[a];
  totMatches += 1;

for b in ind2:
  totGoalsAT += lstFTAG[b];
  totMatches += 1;

for c in ind:
  totalYCHT += lstTYCHT[c];

for d in ind2:
  totalYCAT += lstTYCAT[d];

for e in ind:
  totalRCHT += lstTRCHT[e];

for f in ind2:
  totalRCAT += lstTRCAT[f];

avgGoal = round((totGoalsHT + totGoalsAT) / totMatches, 2);

print("Barcelona total goals in La Liga BBVA 2024/25 as a home team: " + str(totGoalsHT));
print("Barcelona total goals in La Liga BBVA 2024/25 as an away team: " + str(totGoalsAT));
print("Total goals of Barcelona in La Liga BBVA 2024/25: " + str(totGoalsHT + totGoalsAT));
print("Barcelona average goals per match in La Liga BBVA 2024/25: " + str(avgGoal));
print("Barcelona total yellow cards in La Liga 2024/25 season: " + str(totalYCHT + totalYCAT));
print("Barcelona total red cards in La Liga 2024/25 season: " + str(totalRCHT + totalRCAT));
```
